```python
hnl=[]
for year in lonR.index:
    for age_group in houseR.columns:
        correlation_a = lonR.loc[year, age_group]
        correlation_b = houseR.loc[year, age_group]

        if correlation_a == 0:
            continue

        ratio = correlation_b / correlation_a
        hnl.append(round(ratio,2))
        print(f"Year: {year}, Age Group: {age_group}, Ratio: {ratio:.2f}")
```
