```python
#AGE_DVS : 나이
#NOPE : 사람 수
#LON_OWNR_NOPE : 대출
#BANK_LON_OWRN_NOPE : 은행권 대출
#BANK_ETC_LON_OWRN_NOPE : 비은행권 대출
#CARD_LON_OWRN_NOPE : 카드론
```

```python
data={}
data=df[['NOPE','AGE_DVS','LON_OWNR_NOPE','BANK_LON_OWRN_NOPE','BANK_ETC_LON_OWRN_NOPE','CARD_LON_OWRN_NOPE']]
data2=df[['NOPE','AGE_DVS','MM_INCM','CR_AVG']]

data2['CR_AVG'].astype(int)
data2['MM_INCM'].astype(int)
data2['NOPE'].astype(int)
data2.loc[:,'CR_AVG']=data2['CR_AVG']*data2['NOPE']
data2.loc[:, 'MM_INCM'] = data2['MM_INCM'] * data2['NOPE']

data2
```

```python
sum_data=data.groupby('AGE_DVS').sum()
sum_data2=data2.groupby('AGE_DVS').sum()
sum_data2.loc[:,'MM_INCM']=round(sum_data2['MM_INCM'] / sum_data2['NOPE'],2)
sum_data2.loc[:, 'CR_AVG'] = round(sum_data2['CR_AVG'] / sum_data2['NOPE'],2)
sum_data.index=sum_data.index.astype(str)
sum_data2.index=sum_data2.index.astype(str)
sum_data2
```

```python
sum_data['BANK_RATIO']=round(sum_data['BANK_LON_OWRN_NOPE']/sum_data['LON_OWNR_NOPE']*100,3)
sum_data['BANK_ETC_RATIO']=round(sum_data['BANK_ETC_LON_OWRN_NOPE']/sum_data['LON_OWNR_NOPE']*100,3)
sum_data['CARD_RATIO']=round(sum_data['CARD_LON_OWRN_NOPE']/sum_data['LON_OWNR_NOPE']*100,3)
sum_data
```
